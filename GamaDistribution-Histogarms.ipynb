```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.gamma( 2, 3, 100000) #generates random gamma distribution values from 100000 samples
```

```python
plt.hist( x, bins=30)
plt.savefig("normalhistogram.png") 
plt.show() #histogram
```

```python
plt.hist( x, bins=30, normed=True) 
plt.savefig("normalisedhistogram.png") 
plt.show()  #normalised histogram
```

```python
plt.hist( x, bins=30, cumulative = True) 
plt.savefig("cumulativehistogram.png")plt.show()  #cumulative histogram
```

```python
plt.hist( x, bins=30, cumulative = True, normed = True, histtype = "step") 
plt.savefig("norm-cum-histogram.png") 
plt.show()  #both normed and cumulative histogram together
```

```python
plt.figure()
plt.subplot(221)
plt.hist(x,bins=30)
plt.subplot(222)
plt.hist( x, bins=30, normed=True) 
plt.subplot(223)
plt.hist( x, bins=30, cumulative = True) 
plt.subplot(224)
plt.hist( x, bins=30, cumulative = True, normed = True, histtype = "step")
```

```python
plt.savefig("allinone.png") 
plt.show()
```
